```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['dept_hour']=df['Dep_Time'].str.split(':').str[0]
df['dept_min']=df['Dep_Time'].str.split(':').str[1]
df['arr_hour']=df['Arrival_Time'].str.split(':').str[0]
df['arr_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df['arr_min']=df['arr_min'].str.split(' ').str[0]
```

```python
df['dept_hour']=df['dept_hour'].astype(int)
df['dept_min']=df['dept_min'].astype(int)
df['arr_hour']=df['arr_hour'].astype(int)
df['arr_min']=df['arr_min'].astype(int)
```

```python
df.drop(['Dep_Time','Arrival_Time'],axis=1,inplace=True)
```

```python
df['duration_h']=df['Duration'].str.split('h').str[0]
df['duration_m']=df['Duration'].str.split(' ').str[1]
```

```python
df['duration_m']=df['duration_m'].str.split('m').str[0]
```

```python
df['duration_m']=df['duration_m'].fillna(0)
```

```python
df['duration_h']=df['duration_h'].astype(int)
df['duration_m']=df['duration_m'].astype(int)
```

```python
df.drop('Duration',axis=1,inplace=True)
```
